```python
train = train.loc[train["Discount_offered"] <= 10] #reducimos el rango de descuentos al 10%
```

```python
#reemplazamos los valores de cada columna con el de los siguientes diccionarios para un mejor entendimiento
product_importance = {"low":1,"medium":2,"high": 3}
gender = {"F":1, "M":2}
warehouse = {"A":1,"B":2,"C":3,"D":4,"F":5}
shipment = {"Road":1, "Flight":2, "Ship":3}

new_train = train.replace({"Product_importance":product_importance, "Gender":gender, "Warehouse_block":warehouse, 'Mode_of_Shipment': shipment})
new_train
```

```python
#eliminamos la columna ID
new_train.drop(["ID"], axis=1, inplace=True)
new_train.head()
```

```python
#Determinando los valores a utilizar posteriormente
x = new_train.drop("Reached.on.Time_Y.N", axis="columns")
y = new_train["Reached.on.Time_Y.N"]
```

```python
#Adopto un tamaÃ±o de testeo de 30% de los registros totales
#Adopto un numero aleatorio para el random_state, en este caso 10
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10) #Realizamos el split
```

```python
# x_train = np.array(x_train) #Convertimos nuestra variable en un array
# x_train
# x_train = np.array(x_train) #Convertimos nuestra variable en un array
# x_train
```

```python
#Instanciamos el modelo en una variable
logReg = LogisticRegression()
```

```python
#Ajustamos el modelo usando el entrenamiento
logReg = logReg.fit(x_train, y_train)
```

```python
#Usamos el modelo para hacer predicciones del testing
y_pred = logReg.predict(x_test) #Empezamos a pronosticar por el conjunto que no fue considerado(test)
y_pred
```

```python
#Diagnosticos del modelo
conf_mat = metrics.confusion_matrix(y_test, y_pred)
conf_mat
#El primer set diagonal es la pronosticaciÃ³n acertada, y en la segunda son equÃ­vocos.
```

```python
#Evaluamos la calidad del modelo
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```
